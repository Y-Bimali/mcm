```python
import numpy  as np
import trimesh
import open3d as o3d


```

```python

input_file = "untitled.ply"
mesh = trimesh.load(input_file)
print("Displaying mesh...")
mesh.show()

center = mesh.centroid

# Compute the translation along the z-axis
translation = np.array([0, 0, -center[2]])

# Apply the translation to the mesh
mesh.apply_translation(translation)

# Visualize the translated mesh (if you want to use trimesh's built-in viewer)
mesh.show()

```

```python

import csv

# Parameters
num_points = 10000
output_file = f"{input_file[:-4]}-{num_points}pt.csv"

# Sample points from the mesh
points, _ = trimesh.sample.sample_surface(mesh, num_points)

# Display the point cloud
print("Displaying point cloud...")
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = o3d.utility.Vector3dVector(points)
o3d.visualization.draw_geometries([point_cloud])

# Save the point cloud as a CSV file
print(f"Saving point cloud to {output_file}...")
with open(output_file, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["X", "Y", "H"])  # Write header
    csv_writer.writerows(points)         # Write points

print(f"Point cloud saved to {output_file}")
```
